```python
def ratio(dataframe):
    df = dataframe.copy(deep=True)
    df.sort_values(by='Contour_ID',inplace=True)
    
    dfa = df.drop_duplicates(subset='Contour_ID',keep='first',ignore_index=True)
    dfb = df.drop_duplicates(subset='Contour_ID',keep='last',ignore_index=True)
    df = dfa.join(dfb,lsuffix='1',rsuffix='2')
    l = len(df)
    df = df.loc[np.absolute(df['X1']-df['X2'])<3/13.6]
    df = df.loc[np.absolute(df['Y1']-df['Y2'])<3/13.6]
    m = len(df)
    
    return m/l
```

```python
def pltrend(group,num):
    dataframes = pd.DataFrame()
    for dataframe in group:
        df = dataframe.copy(deep=True)
        df.sort_values(by='Contour_ID',inplace=True)
        dfa = df.drop_duplicates(subset='Contour_ID',keep='first',ignore_index=True)
        dfb = df.drop_duplicates(subset='Contour_ID',keep='last',ignore_index=True)
        dfd = dfa.join(dfb,lsuffix='1',rsuffix='2')
        dataframes = pd.concat([dataframes,dfd.drop_duplicates()])
    dfs = dataframes.drop_duplicates(subset=['Contour_ID1','Contour_ID2'],ignore_index=True)
    
    df = pd.DataFrame()
    for i in group:
        df = pd.concat([df,i])
     
    trends = []
    itv = np.arange(0,2.5/13.6,0.5/13.6)
    for j in itv:
        dfc = dfs.loc[np.absolute(dfs['X1']-dfs['X2'])>j]
        dfc = dfc.loc[np.absolute(dfc['Y1']-dfc['Y2'])>j]
        dfc = dfc.loc[np.absolute(dfc['X1']-dfc['X2'])<=j+0.5/13.6]
        dfc = dfc.loc[np.absolute(dfc['Y1']-dfc['Y2'])<=j+0.5/13.6]

        a = np.array(dfc.Contour_ID1.unique())
        if len(a)==0:
            trends.append(0)
            break
        if len(a)<=num:
            num = len(a)
        b = np.random.randint(len(a),size=num)
        c = np.array(a[b])

        result = pd.DataFrame()
        for i in c:            
            d = df.query('Contour_ID == @i')
            ang = np.array(d['Aon'])
            inv = ang.reshape([1,-1]).T
            cos = np.cos(ang-inv).flatten()

            l = len(ang)
            x = np.arange(l)
            y = x.reshape([1,-1]).T
            r = np.absolute((x-y).flatten())

            e = pd.DataFrame({'r':r/13.6,'cos':cos}).groupby('r').mean() 
            result = pd.concat([result,e]) 
        result = result.groupby('r').mean()

        a,b = np.polyfit(result.index[0:18],np.log(result.cos).head(18),1)
        trends.append(-1/a)

    return trends
```
