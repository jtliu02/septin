```python
#calculate alignments

#low density
align1 = alignmentM(l1,100000)
align2 = alignmentM(l2,100000)
align3 = alignmentM(l3,100000)
align4 = alignmentM(l4,100000)
align5 = alignmentM(l5,100000)
#medium density
align14 = alignmentM(m14,100000)
align15 = alignmentM(m15,100000)
align17 = alignmentM(m17,100000)
align18 = alignmentM(m18,100000)
align19 = alignmentM(m19,100000)
#high density
align24 = alignmentM(h24,100000)
align25 = alignmentM(h25,100000)
align26 = alignmentM(h26,100000)
align27 = alignmentM(h27,100000)
align28 = alignmentM(h28,100000)
#gic
aligng12 = alignmentM(g12,100000)
aligng13 = alignmentM(g13,100000)
#cla
alignc28 = alignmentM(c28,100000)
alignc29 = alignmentM(c29,100000)

#to save results:
#align1.to_csv('align1.csv')

#to read saved results:
#align1 = pd.read_csv('align1.csv')
```

```python
#sample alignment dataframe
align1
```

```python
#alignment arrays

lowA = pd.concat([align1,align2,align3,align4,align5]).sort_values(by=['r'])
medA = pd.concat([align14,align15,align17,align18,align19]).sort_values(by=['r'])
highA = pd.concat([align24,align25,align26,align27,align28]).sort_values(by=['r'])
gicA = pd.concat([aligng12,aligng13]).sort_values(by=['r'])
claA = pd.concat([alignc28,alignc29]).sort_values(by=['r'])

alignments = [lowA,medA,highA,gicA,claA]

lowA
```

```python
#nematic correlation length

nematic = []
bn = []
cin = []

plt.figure(figsize=(5,5))
for i in alignments:
    
    #scatter graph
    y0 = i.groupby('r')['align'].mean() #groupby r
    idx0 = y0[y0>0].index #find all indices in which alignment is positive
    y1 = np.log(y0[idx0]) #take log of all those positive alignment
    plt.scatter(idx0,y1,alpha=0.5) #graph log(align) vs r
    
    #compute fit
    y2 = i.groupby('r')['align'].mean().head(14) #change 38 (1.2 micron) to change range at which we fit the function
    idx1 = y2[y2>0].index
    y = np.log(y2[idx1])
    p,cov = np.polyfit(idx1,y,1,cov=True) #fit first degree polynomial by the first 14 points
    
    #graph fitted curve
    x = np.array([0,10])
    plt.plot(x,p[0]*x+p[1]) #plot fit
    
    #document data  
    nematic.append(-1/p[0]) #nematic = -1/slope
    bn.append(p[1]) #y = mx + b. b value
    cin.append(2*np.sqrt(cov[0][0]/len(idx1))) #95% CI = 2*SE = 2*SD/sqrt(n) = 2*sqrt(variance/n)
    
plt.xlim(left=0,right=10)
plt.ylim(bottom=-10,top=0)

plt.xlabel('r (micron)')
plt.ylabel('log(alignment)');
```

```python
#alignment vs r

plt.figure(figsize=(5,5))
a = 0
x = np.linspace(0,10,100)
for i in alignments:
    plt.scatter(i.r.unique(),i.groupby('r')['align'].mean(),label=name[a],alpha=0.3)
    plt.plot(x,np.e**(-x/nematic[a]+bn[a]))
    a += 1
plt.legend(loc=0)
plt.xlabel('r (micron)')
plt.ylabel('alignment')
plt.plot([0,10],[0,0]);
```

```python
#compare nematic order in bar graph (with density)

for i in range(len(name)):
    den = density[i]
    nem = nematic[i]
    ci_s = cin[i]
    lsem = 1/(1/np.array(nem)-np.array(ci_s))-np.array(nem)
    usem = 1/(1/np.array(nem)+np.array(ci_s))-np.array(nem)
    
    plt.bar(den,nem,width=0.4*10**8,color='C'+str(i),label=name[i],alpha=0.7)
    plt.legend(loc='best')
    plt.errorbar(den,nem,yerr=[[-lsem],[usem]],ecolor='black',capsize=10,alpha=0.7)
    
plt.xlabel('density')
plt.ylabel('nematic order');
```

```python
#compare nematic order in bar graph (without density)

for i in range(len(name)):
    nem = nematic[i]
    ci_s = cin[i]
    lsem = 1/(1/np.array(nem)-np.array(ci_s))-np.array(nem)
    usem = 1/(1/np.array(nem)+np.array(ci_s))-np.array(nem)
    
    plt.bar(i,nem,color='C'+str(i),alpha=0.7)
    plt.errorbar(i,nem,yerr=[[-lsem],[usem]],ecolor='black',capsize=10,alpha=0.7)

plt.xlabel('category')
plt.ylabel('nematic order')
plt.xticks(np.arange(len(name)),name);
```

```python
lowP = pd.concat([pl1,pl2,pl3,pl4,pl5]).sort_values(by='r')
medP = pd.concat([pl14,pl15,pl17,pl18,pl19]).sort_values(by='r')
highP = pd.concat([pl24,pl25,pl26,pl27,pl28]).sort_values(by='r')
gicP = pd.concat([plg12,plg13]).sort_values(by='r')
claP = pd.concat([plc28,plc29]).sort_values(by='r')

pls = [lowP,medP,highP,gicP,claP]

lowP
```

```python
#persistence length

persistence = []
bp = []
cip = []

plt.figure(figsize=(5,5))
for i in pls:

    #scatter graph
    y0 = i.groupby('r')['cos'].mean() #groupby r
    idx0 = y0[y0>0].index #find all indices in which persistence is positive
    y1 = np.log(y0[idx0]) #take log of all those positive persistence
    plt.scatter(idx0,y1,alpha=0.5) #graph log(persistence) vs r
    
    #compute fit
    y2 = i.groupby('r')['cos'].mean().head(14) #change to change range at which we fit the function
    idx1 = y2[y2>0].index
    y = np.log(y2[idx1])
    p,cov = np.polyfit(idx1,y,1,cov=True) #fit first degree polynomial by the first 14 points
    
    #graph fitted curve
    x = np.array([0,10])
    plt.plot(x,p[0]*x+p[1]) #plot fit
    
    #document data  
    persistence.append(-1/p[0]) #persistence = -1/slope
    bp.append(p[1]) #y = mx + b. b value
    cip.append(2*np.sqrt(cov[0][0]/len(idx1))) #95% CI = 2*SE = 2*SD/sqrt(n) = 2*sqrt(variance/n)

plt.ylim(bottom=-1.5,top=0)
plt.xlabel('r (micron)')
plt.ylabel('log(flex)');

#warnings due to taking log of negative. not to be worried
```

```python
#flexibility vs r

plt.figure(figsize=(5,5))
a = 0
x = np.linspace(0,15,150)
for i in pls:
    plt.scatter(i.r.unique(),i.groupby('r')['cos'].mean(),label=name[a],alpha=0.3)
    plt.plot(x,np.e**(-x/persistence[a]+bp[a]))
    a += 1
plt.legend(loc=0)
plt.xlabel('r (micron)')
plt.ylabel('flex')
plt.plot([0,15],[0,0]);
```

```python
#compare flex in bar graph (with density)

for i in range(len(name)):
    den = density[i]
    per = persistence[i]
    ci_s = cip[i]
    lsem = 1/(1/np.array(per)-np.array(ci_s))-np.array(per)
    usem = 1/(1/np.array(per)+np.array(ci_s))-np.array(per)
    
    plt.bar(den,per,width=0.4*10**8,color='C'+str(i),label=name[i],alpha=0.7)
    plt.legend(loc='best')
    plt.errorbar(den,per,yerr=[[-lsem],[usem]],ecolor='black',capsize=10,alpha=0.7)
    
plt.xlabel('density')
plt.ylabel('flex');
```

```python
#compare flex in bar graph (without density)

for i in range(len(name)):
    per = persistence[i]
    ci_s = cip[i]
    lsem = 1/(1/np.array(per)-np.array(ci_s))-np.array(per)
    usem = 1/(1/np.array(per)+np.array(ci_s))-np.array(per)
    
    plt.bar(i,per,color='C'+str(i),alpha=0.7)
    plt.errorbar(i,per,yerr=[[-lsem],[usem]],ecolor='black',capsize=10,alpha=0.7)

plt.xlabel('category')
plt.ylabel('flex')
plt.xticks(np.arange(len(name)),name);
```
